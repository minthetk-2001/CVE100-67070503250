You bought 500 shares of stock A at $\$$600 on day 0 and you sold it at  $\$$1000 on day 700.
The daily discount rate is 0.0001%. How much profit did you make in terms of net present value.


```python
#Write your code here

#Given
shares = 500
buy_price = 600
sell_price = 1000
days = 700
daily_discount_rate = 0.000001  # 0.0001% in decimal

initial_outflow = shares * buy_price
future_inflow = shares * sell_price

# Present value of inflow
pv_inflow = future_inflow / ((1 + daily_discount_rate) ** days)

# Net present value
npv_profit = pv_inflow - initial_outflow
print(("Profit made in terms of net present value is"), npv_profit)
```
